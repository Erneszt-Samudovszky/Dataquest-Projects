# Project: Visualizing Earnings Based On College Majors

In this project we will work pandas plotting functionality in order to visualize and analyze our dataset, which represents job outcomes of students who graduated from college between 2010 and 2012.

Each row in the dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries, and more. Here are some of the columns in the dataset:

   *  Rank - Rank by median earnings (the dataset is ordered by this column).
   *  Major_code - Major code.
   *  Major - Major description.
   *  Major_category - Category of major.
   *  Total - Total number of people with major.
   *  Sample_size - Sample size (unweighted) of full-time.
   *  Men - Male graduates.
   *  Women - Female graduates.
   *  ShareWomen - Women as share of total.
   *  Employed - Number employed.
   *  Median - Median salary of full-time, year-round workers.
   *  Low_wage_jobs - Number in low-wage service jobs.
   *  Full_time - Number employed 35 hours or more.
   *  Part_time - Number employed less than 35 hours.
 
Using visualizations, we can start to explore questions from the dataset like:

   *  Do students in more popular majors make more money?
       *  Using scatter plots
   *  How many majors are predominantly male? Predominantly female?
       *  Using histograms
   *  Which category of majors have the most students?
       *  Using bar plots

## Pandas, Scatter Plots

We can start plotting and analyzing the data.

```python
ax = recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
ax.set_title('Sample_size vs. Median')
```

```python
ax = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
ax.set_title('Sample size vs. Unemployment_rate')
```

```python
ax = recent_grads.plot(x='Full_time', y='Median', kind='scatter')
ax.set_title('Full_time vs. Median')
```

The plot shows that the vast majority of the graduates employed full-time have a median salary between 30000 and 45000.

```python
ax = recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
ax.set_title('ShareWomen vs. Unemployment_rate')
```

There is a really weak correlation here.

```python
ax = recent_grads.plot(x='Men', y='Median', kind='scatter')
ax.set_title('Men vs. Median')
```

```python
ax = recent_grads.plot(x='Women', y='Median', kind='scatter')
ax.set_title('Women vs. Median')
```

## Pandas, Bar plots

Now we can proceed further and visualize bar plots to compare the percentages of women (**ShareWomen**) from the first then rows and last ten rows of our dataset.

```python
recent_grads[:10].plot(kind='bar', x='Major', y='ShareWomen')
recent_grads[163:].plot(kind='bar', x='Major', y='ShareWomen')
```

Looks like enginnering majors are mostly occupied by men, although Astronomy and Astrophysics is somewhat balanced. **Early Childhood Education**, **Other Foreign languages**, **Drama and Theater Arts**, **Composition and Rhetoric**, **Zoology**, **Educational Psychology**, **Clinical Psychology**, **Counseling Psychology** and **Library Science** majors are mostly chosen by women.

We can try to do the same for the unemployment rate (**Unemployment_rate**).

```python
recent_grads[:10].plot(kind='bar', x='Major', y='Unemployment_rate')
recent_grads[163:].plot(kind='bar', x='Major', y='Unemployment_rate')
```

We notice that our results vary quite a bit. The biggest numbers for the unemployment rate are for **Nuclear Engineering** and **Clinical Psychology**. On the other hand, **Petroleum Engineering**, **Metallurgical Engineering** and **Astronomy and Astrophysics** majors show the lowest unemployment rate data.
